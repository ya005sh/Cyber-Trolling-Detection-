```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
df = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    '"#vaccine"').get_items(),100))
print(df.head())
```

```python
data = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    '"#vaccine since:2021-05-01 until:2021-07-30"').get_items(), 50000))
```

```python
data.head()
```

```python
df = pd.DataFrame(data,columns=['date','content','username'])
```

```python
df.head()
```

```python
df.shape
```

```python
# saving the dataframe
df.to_csv('/content/drive/MyDrive/Colab Notebooks/Cybertrolling_Project/Scrapper_data1.csv')
```
