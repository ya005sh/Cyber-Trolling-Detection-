```python
X = df['comment_text']
y = df[df.columns[2:]].values
```

```python
MAX_FEATURES = 200000
```

```python
vectorizer = TextVectorization(max_tokens=MAX_FEATURES,output_sequence_length=1800,
                                output_mode='int')
```

```python
vectorizer.adapt(X.values)
```

```python
vectorizer('Hello world , life is beautiful')[:5]
```

```python
vectorized_text = vectorizer(X.values)
```

```python
vectorized_text
```

```python
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8)
```

```python
batch_X,batch_y = dataset.as_numpy_iterator().next()
```

```python
len(dataset) # not actual size this is batch size
```

```python
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1)) 
```

```python
train_generator = train.as_numpy_iterator()
```

```python
train_generator.next()
```

## Creating Sequential Model 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dropout,Bidirectional,Dense,Embedding
```

```python
model = Sequential()

model.add(Embedding(MAX_FEATURES+1,32))
model.add(Bidirectional(LSTM(32,activation='tanh')))
model.add(Dense(128,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))

model.add(Dense(6,activation='sigmoid'))


```

```python
model.compile(loss='BinaryCrossentropy',optimizer='Adam')
```

```python
model.summary()
```

```python
history = model.fit(train,epochs=1,validation_data=val)
```

```python
from tensorflow.keras.metrics import Precision,Recall,CategoricalAccuracy
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X_true,y_true = batch
    yhat = model.predict(X_true)

    y_true = y_true.flatten()
    yhat = yhat.flatten()

    pre.update_state(y_true,yhat)
    re.update_state(y_true,yhat)
    acc.update_state(y_true,yhat)
```

```python
print(f'Precision:{pre.result().numpy()}, Recall:{pre.result().numpy()},Accuracy:{acc.result().numpy()}')
```
